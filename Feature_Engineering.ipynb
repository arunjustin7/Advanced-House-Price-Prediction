```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
pd.pandas.set_option("display.max_columns", None)
```

```python
dataset = pd.read_csv("train.csv")
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train, y_test = train_test_split(dataset,dataset['SalePrice'], test_size = 0.25, random_state= 0)
```

```python
nan_values = [feature for feature in dataset.columns if dataset[feature].isnull().sum() >1 and dataset[feature].dtype == "O"]
for feature in nan_values:
    print(" {} : {}% missing values".format(feature, np.round(dataset[feature].isnull().mean(),4)))
```

```python
def replace_cat (dataset, nan_values):
  data = dataset.copy()
  data[nan_values]= data[nan_values].fillna("missing")
  return data
dataset= replace_cat(dataset, nan_values)
dataset[nan_values].isnull().sum()
```

```python
(dataset[nan_values] == "missing").sum()
```

```python
numerical_features = [feature for feature in dataset.columns if dataset[feature].isnull().sum() >1 and dataset[feature].dtype != "0"]

for feature in numerical_features :
  print(" {}: {}% missing values ".format(feature, np.round(dataset[feature].isnull().mean(),4)))
```

```python
for feature in numerical_features:
  median_value = dataset[feature].median()
  dataset[feature + 'nan'] = np.where(dataset[feature].isnull(),1,0)
  dataset[feature].fillna(median_value, inplace=True)
dataset[numerical_features
        ].isnull().sum()
```

```python
for feature in dataset[['YearBuilt','YearRemodAdd','GarageYrBlt']]:
  dataset[feature] = dataset['YrSold'] - dataset[feature]
```

```python
dataset[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

```python
num_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
  dataset[feature]= np.log(dataset[feature])
```

```python
dataset[num_features].head()
```

```python
cat_features = [ feature  for feature in dataset.columns if dataset[feature].dtype == 'O']
```

```python
cat_features
```

```python
for feature in cat_features:
  temp = dataset.groupby(feature)['SalePrice'].count()/len(dataset)
  temp_df =temp[temp >0.01].index
  dataset[feature] = np.where(dataset[feature].isin(temp_df),dataset[feature], 'Rare_VAR')
```

```python
dataset.head(100)
```

```python
for feature in cat_features:
  labels_ordered = dataset.groupby(feature)['SalePrice'].mean().sort_values().index
  labels_ordered = { k: i for i,k in enumerate(labels_ordered,0) }
  dataset[feature]= dataset[feature].map(labels_ordered)
```

```python
dataset.head()
```

```python
numeric_features = [feature for feature in dataset.columns if feature not in ["Id","SalePrice"]]

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(dataset[numeric_features])
```

```python
data = pd.concat([dataset[["Id","SalePrice"]].reset_index(drop=True),pd.DataFrame(scaler.transform(dataset[numeric_features]),columns=numeric_features)],axis=1)
```

```python
data.head()
```

```python
data.to_csv('X_train.csv', index=False)
```
