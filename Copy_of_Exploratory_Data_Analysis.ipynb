```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
```

```python
dataset = pd.read_csv("train.csv")
dataset.shape
```

```python
dataset.head()
```

# **misssing values**

```python
feature_with_na = [features for features in dataset.columns if dataset[features].isnull().sum() >1]

for feature in feature_with_na:
  print(feature, np.round(dataset[feature].isnull().mean(),4), '% missing values')
```

```python
for feature in feature_with_na:
  data = dataset.copy()
  data[feature] = np.where(data[feature].isnull(), 1, 0)
  data.groupby(feature)['SalePrice'].median().plot.bar()
  plt.title(feature)
  plt.show()
```

```python
print(" count of the id {}".format(len(dataset['Id'])))
```

# **NUMERICAL VARIABLES**

```python
numerical_values = [features for features in dataset.columns if dataset[features].dtypes != "O"]
print("the number of num values are : ", len(numerical_values))
dataset[numerical_values].head()
```

# **date and time numerical value**

```python
year_feature = [feature for feature in numerical_values if "Yr" in feature or "Year" in feature]
year_feature
```

```python
for feature in year_feature:
  print(feature, dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel("YEAR SOLD")
plt.ylabel("SALES PRICE")
plt.title("SALES PRICE vs YEAR SOLD")
```

```python
for feature in year_feature:
  if feature != "YrSold":
    data = dataset.copy()
    data[feature] = data["YrSold"] - data[feature]

    plt.scatter(data[feature], data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel("SalePrice")
    plt.title("difff btwn saleprice and feature")
    plt.show()
```

```python
descrete_feature = [feature for feature in numerical_values if len(dataset[feature].unique())<25 and feature not in year_feature + ['Id'] ]
print("the count of descrete features is:{}".format(len(descrete_feature)))
```

```python
descrete_feature
```

```python
dataset[descrete_feature].head()
```

```python
for feature in descrete_feature:
  data = dataset.copy()
  data.groupby(feature)['SalePrice'].median().plot.bar()
  plt.xlabel(feature)
  plt.ylabel("descrete feature")
  plt.show()
```

```python
continuous_feature = [feature for feature in numerical_values if feature not in descrete_feature + year_feature + ['Id']]
print("the count of the continous feature is: {}".format(len(continuous_feature)))
```

```python
for faeture in continuous_feature:
  data =dataset.copy()
  data[feature].hist(bins = 25)
  plt.xlabel(feature)
  plt.ylabel("count")
  plt.title("Count of cn features")
  plt.show()
```

```python
for feature in continuous_feature:
  data = dataset.copy()
  if 0 in data[feature].unique():
    pass
  else:
    data[feature] = np.log(data[feature])
    data['SalePrice']= np.log(data["SalePrice"])
    plt.scatter(data[feature], data["SalePrice"])
    plt.xlabel(feature)
    plt.ylabel("SalePrice")
    plt.title(feature)
    plt.show()
```

```python
for feature in continuous_feature:
  data = dataset.copy()
  if 0 in data[feature].unique():
    pass
  else:
    data[feature] = np.log(data[feature])
    data.boxplot(column=feature)
    plt.ylabel(feature)
    plt.title(feature)
    plt.show()
```
